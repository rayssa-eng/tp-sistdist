```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(x, y):
    # Create a figure and axes
    fig, ax = plt.subplots()

    # Plot the data
    ax.scatter(x, y)
    
    # Add labels and title
    ax.set_xlabel('numero de processos')
    ax.set_ylabel('tempo (s)')
    ax.set_title('Tempo para a execução de todos os clientes')

    # Set explicit x-axis tick labels
    ax.set_xticks(x)
    ax.plot(x, y, color='red', linestyle='--', linewidth=0.2)

    # Display the graph
    plt.show()
```

```python
x_1 = [2, 4, 8, 16, 32]
y_1 = [38.037, 78.082, 158.164, 318.323, 638.642]

x_2 = [2, 4, 8, 16, 32, 64]
y_2 = [4.007, 4.013, 4.024, 4.05, 4.116, 4.303]

x_3 = [2, 4, 8, 16, 32, 64, 128]
y_3 = [0.86, 0.88, 0.104, 0.131, 0.200, 0.471, 0.849]
```

```python
plot(x_1, y_1)
plot(x_2, y_2)
plot(x_3, y_3)
```
